# Making dataframes to be used for each Use Case 

5 datasets:
- data-1.json => Private Residential Property Transactions
- data-2.json => Private Non-Landed Residential Properties Median Rentals by Name

# For Private Residential Properties Transactions

This data service will return past 5 years of private residential property transaction records.


```python
pvtResidential = pd.read_csv('data-1.csv')
pvtResidential.head()
```

```python
pvtResidential.info()
```

```python
#Cleaning data

pvtResidential.fillna(0, inplace = True)

#Dropping irrelevant columns

newpvtResidential = pvtResidential.drop(['Status','Result.transaction.area','Result.transaction.nettPrice','Result.transaction.floorRange','Result.transaction.district','Result.transaction.contractDate','Result.transaction.typeOfSale', 'Result.x', 'Result.project', 'Result.y', 'Result.marketSegment'], axis = 1)

print(newpvtResidential.head())
```

```python
newpvtResidential.info()
```

```python
newpvtResidential.rename(columns = {'Result.transaction.price':'Price'}, inplace = True)
newpvtResidential.rename(columns = {'Result.street':'Street'}, inplace = True)
newpvtResidential.rename(columns = {'Result.transaction.propertyType':'Property Type'}, inplace = True)
newpvtResidential.rename(columns = {'Result.transaction.typeOfArea':'Type of Area'}, inplace = True)
newpvtResidential.rename(columns = {'Result.transaction.tenure':'Transaction Tenure'}, inplace = True)
newpvtResidential.rename(columns = {'Result.transaction.noOfUnits':'No.of Units Available'}, inplace = True)
newpvtResidential.head()
```

```python
newpvtResidential.info()
```

# For HDB Flats

```python
hdbDetails = pd.read_csv('data-5-govsg.csv')
hdbDetails.head()
```

```python
#Cleaning data

hdbDetails.fillna(0, inplace = True)

#Renaming Column Headings

hdbDetails.rename(columns = {'/_id':'ID'}, inplace = True)
hdbDetails.rename(columns = {'/block':'Block No.'}, inplace = True)
hdbDetails.rename(columns = {'/flat_model':'Flat Type'}, inplace = True)
hdbDetails.rename(columns = {'/floor_area_sqm':'Square Meters'}, inplace = True)
hdbDetails.rename(columns = {'/lease_commence_date':'Lease Start Date'}, inplace = True)
hdbDetails.rename(columns = {'/month':'Month'}, inplace = True)
hdbDetails.rename(columns = {'/remaining_lease':'Remaining Years of Lease'}, inplace = True)
hdbDetails.rename(columns = {'/resale_price':'Price'}, inplace = True)
hdbDetails.rename(columns = {'/storey_range':'Storey Range'}, inplace = True)
hdbDetails.rename(columns = {'/street_name':'Street Name'}, inplace = True)
hdbDetails.rename(columns = {'/town':'Area'}, inplace = True)
hdbDetails.head()
```

#  Private Residential Projects in the Pipeline

This refers to residential projects that are in the process of being planned or developed

This can come under the NEW PROJECTS section in our website

```python
projPipeline = pd.read_csv('data-4.csv')
projPipeline.head()
```

```python
#Cleaning data

projPipeline.fillna(0, inplace = True)

#Dropping irrelevant columns

projPipeline = projPipeline.drop(['Result.expectedTOPYear'], axis = 1)
print(projPipeline.head())
```

```python
#Renaming Columns

projPipeline.rename(columns = {'Result.noOfApartment':'No. of Apartments'}, inplace = True)
projPipeline.rename(columns = {'Result.noOfSemiDetached':'No. of Semi-Detached'}, inplace = True)
projPipeline.rename(columns = {'Result.street':'Street Name'}, inplace = True)
projPipeline.rename(columns = {'Result.district':'District Name'}, inplace = True)
projPipeline.rename(columns = {'Result.project':'Project Name'}, inplace = True)
projPipeline.rename(columns = {'Result.noOfDetachedHouse':'No. of Detached Houses'}, inplace = True)
projPipeline.rename(columns = {'Result.noOfTerrace':'No. of Houses with Terrace'}, inplace = True)
projPipeline.rename(columns = {'Result.noOfCondo':'No. of Condo'}, inplace = True)
projPipeline.rename(columns = {'Result.developerName':'Developer Name'}, inplace = True)
projPipeline.rename(columns = {'Result.totalUnits':'Total No. of Units'}, inplace = True)


projPipeline.head()
```

# Private Residential Property Units Sold by Developers

This data service will return past 3 years of prices of completed and uncompleted private residential units and executive condominiums with pre-requisite for sale sold by developers


Divided into 3 datasets by refPeriod (data that is available for download by reference quarter):
1. data-3 => refPeriod: 0123
2. data-3-1 => refPeriod: 1222
3. data-3-2 => refPeriod: 1122


```python
#dataset data-3

unitSold = pd.read_csv("data-3.csv")
unitSold.head()
```

```python
#Cleaning data

unitSold.fillna(0, inplace = True)

#Dropping irrelevant columns

unitSold = unitSold.drop(['Result.district', 'Result.x', 'Result.y', 'Result.developerSales.refPeriod'], axis = 1)
unitSold.head()
```

```python
unitSold = unitSold.drop(['Result.developer','Result.developerSales.launchedToDate','Result.developerSales.soldToDate', 'Result.developerSales.launchedInMonth','Result.developerSales.soldInMonth', 'Result.marketSegment'],axis=1)
unitSold.head()
```

```python
#Renaming Columns

unitSold.rename(columns = {'Result.street':'Street Name'}, inplace = True)
unitSold.rename(columns = {'Result.propertyType':'Property Type'}, inplace = True)
unitSold.rename(columns = {'Result.project':'Project Name'}, inplace = True)
unitSold.rename(columns = {'Result.developerSales.medianPrice':'Median Price'}, inplace = True)
unitSold.rename(columns = {'Result.developerSales.highestPrice':'Highest Price'}, inplace = True)
unitSold.rename(columns = {'Result.developerSales.lowestPrice':'Lowest Price'}, inplace = True)
unitSold.rename(columns = {'Result.developerSales.unitsAvail':'Units Available'}, inplace = True)

unitSold.head()
```
